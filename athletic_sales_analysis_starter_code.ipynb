```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020 = pd.read_csv("Resources/athletic_sales_2020.csv")
athletic_sales_2021 = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
print(athletic_sales_2020.dtypes)
```

```python
# Check the 2021 sales data types.
print(athletic_sales_2021.dtypes)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_both = pd.concat([athletic_sales_2020,athletic_sales_2021],ignore_index = True)

# Directions say to use an inner join with an index reset, but I think this
# concat function is what is really being asked for.

#also check to make sure this worked:
athletic_sales_both
```

```python
# Check if any values are null.
print(athletic_sales_both.isnull().any())

```

```python
# Check the data type of each column
print(athletic_sales_both.dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_both["invoice_date"] = athletic_sales_both["invoice_date"].apply(pd.to_datetime)
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(athletic_sales_both.dtypes)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

grouped_athletic_sales = (athletic_sales_both.groupby(
    ["region","state","city"])["units_sold"].sum()
    .reset_index(name='Total_Products_Sold'))


# Show the top 5 results.
sorted_athletic_sales = grouped_athletic_sales.sort_values("Total_Products_Sold",
                                                           ascending=False)
sorted_athletic_sales.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table = pd.pivot_table(athletic_sales_both, index=['region', 'state', 'city']
                             , values='units_sold', aggfunc='sum')


# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table = pivot_table.rename(columns = {'units_sold': 'Total_Products_Sold'})
# Show the top 5 results.
sorted_pivot_table = pivot_table.sort_values('Total_Products_Sold',ascending = False)
sorted_pivot_table.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

grouped_athletic_sales2 = (athletic_sales_both.groupby(
    ["region","state","city"])["total_sales"].sum()
    .reset_index(name='Total Sales'))


# Show the top 5 results.
sorted_athletic_sales2 = grouped_athletic_sales2.sort_values("Total Sales",
                                                             ascending=False)
sorted_athletic_sales2.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

pivot_table2 = pd.pivot_table(athletic_sales_both, index=['region', 'state', 'city']
                             , values='total_sales', aggfunc='sum')


pivot_table2 = pivot_table2.rename(columns = {'total_sales': 'Total Sales'})
# Show the top 5 results.
sorted_pivot_table2 = pivot_table2.sort_values('Total Sales',ascending = False)
sorted_pivot_table2.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
athletic_sales_women_shoes = (athletic_sales_both[athletic_sales_both['product']
                                                  == "Women's Street Footwear"])
athletic_sales_women_shoes
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
athletic_sales_women_shoes_grouped = athletic_sales_women_shoes.groupby([
    "retailer","region","state","city"])["units_sold"].sum().reset_index(name =
    "Womens_Footwear_Units_Sold")
# Show the top 5 results.
athletic_sales_women_shoes_grouped = athletic_sales_women_shoes_grouped.sort_values(
    "Womens_Footwear_Units_Sold",ascending = False)
athletic_sales_women_shoes_grouped.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
women_shoe_pivot = pd.pivot_table(athletic_sales_women_shoes, index = [
    "retailer", "region", "state","city"],values = "units_sold",aggfunc = "sum")

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_shoe_pivot = women_shoe_pivot.rename(columns = {"units_sold" : "Womens_Footwear_Units_Sold"})
# Show the top 5 results.
women_shoe_pivot = women_shoe_pivot.sort_values("Womens_Footwear_Units_Sold", ascending = False)
women_shoe_pivot.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
women_shoes_sales_date_pivot = pd.pivot_table(athletic_sales_women_shoes
                                              , index = "invoice_date"
                                              , values = "total_sales"
                                              , aggfunc = "sum")

# Optional: Rename the "total_sales" column to "Total Sales"
women_shoes_sales_date_pivot = women_shoes_sales_date_pivot.rename(columns = {"total_sales":"Total Sales"})

# Show the table.
women_shoes_sales_date_pivot
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_pivot = women_shoes_sales_date_pivot.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_pivot = daily_pivot.sort_values("Total Sales",ascending = True)
daily_pivot

# the directions say to sort in ascending order, but also to show the top 10,
# so it was probably meant to say that we should sort in descending order.
# because of that, I will do that here also:

daily_pivot = daily_pivot.sort_values("Total Sales",ascending = False)
daily_pivot
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_pivot = women_shoes_sales_date_pivot.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_pivot = weekly_pivot.sort_values("Total Sales",ascending = False)
weekly_pivot
```
